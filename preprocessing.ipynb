```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from imblearn.under_sampling import RandomUnderSampler
```

```python
dataset = pd.read_csv("data2023.csv")
```

```python
# Plot distribusi kelas sebelum RUS
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
dataset['k2q31a'].value_counts().plot(kind='bar', title='Distribusi Kelas')
plt.xlabel('Kelas')
plt.ylabel('Jumlah')
plt.xticks([0, 1], ['Tidak ADHD', 'ADHD'])
plt.show()

# Menghitung frekuensi untuk variabel 'concussion' berdasarkan ADHD
freq_concussion = dataset.groupby('k2q31a')['memorycond'].value_counts().unstack(fill_value=0)
categories_concussion = freq_concussion.columns  
category_labels = {1: 'Ya', 2: 'Tidak'}
x = range(len(freq_concussion.index))  
bar_width = 0.2  
plt.figure(figsize=(10, 5))
for i, category in enumerate(categories_concussion):
    plt.bar([pos + i * bar_width - (len(categories_concussion) * bar_width) / 2 for pos in x], 
            freq_concussion[category], width=bar_width, 
            label=category_labels.get(category, category), 
            edgecolor='black', alpha=0.7)
plt.title('Distribusi variabel k2q33a Berdasarkan Label ADHD', fontsize=14)
plt.xticks(x, ['Tidak ADHD', 'ADHD'], fontsize=12)
plt.ylabel('Frekuensi', fontsize=12)
plt.legend(title='Apakah anak kesulitan untuk konsentrasi, mengingat, dan membuat keputusan', fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# Menghitung frekuensi untuk variabel 'concussion' berdasarkan ADHD
freq_concussion = dataset.groupby('k2q31a')['memorycond'].value_counts().unstack(fill_value=0)
categories_concussion = freq_concussion.columns  
category_labels = {1: 'Ya', 2: 'Tidak'}
x = range(len(freq_concussion.index))  
bar_width = 0.2  
plt.figure(figsize=(10, 5))
for i, category in enumerate(categories_concussion):
    plt.bar([pos + i * bar_width - (len(categories_concussion) * bar_width) / 2 for pos in x], 
            freq_concussion[category], width=bar_width, 
            label=category_labels.get(category, category), 
            edgecolor='black', alpha=0.7)
plt.title('Distribusi variabel k2q33a Berdasarkan Label ADHD', fontsize=14)
plt.xticks(x, ['Tidak ADHD', 'ADHD'], fontsize=12)
plt.ylabel('Frekuensi', fontsize=12)
plt.legend(title='Apakah anak kesulitan untuk konsentrasi, mengingat, dan membuat keputusan', fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# Membuat boxplot usia anak berdasarkan ADHD
dataset['age'] = 2023 - dataset['birth_yr']
plt.figure(figsize=(10, 5))
sns.boxplot(x='k2q31a', y='age', data=dataset, palette='pastel')
plt.title('Boxplot Usia Anak', fontsize=14)
plt.xticks([0, 1], ['Tidak ADHD', 'ADHD'], fontsize=12)
plt.ylabel('Usia (tahun)', fontsize=12)
plt.xlabel('ADHD', fontsize=12)
plt.tight_layout()
plt.show()

# Boxplot boxplot usia ibu Berdasarkan ADHD
plt.figure(figsize=(10, 5))
sns.boxplot(x='k2q31a', y='a1_age', data=dataset, palette='pastel')
plt.title('Boxplot Usia Ibu', fontsize=14)
plt.xticks([0, 1], ['Tidak ADHD', 'ADHD'], fontsize=12)
plt.ylabel('Usia ibu (tahun)', fontsize=12)
plt.tight_layout()
plt.show()
```
